# Retention Rates

```python
df['cohort date'] = df.groupby('EmailID')['Order Date'].transform("min")
```

```python
# def func(row):
#     if row['month'] <= 6:
#         return pd.Timestamp(row['Order year'], 1, 1)
#     else:
#         return pd.Timestamp(row['Order year'], 7, 1)
        
    
# df['month'] = df['cohort date'].dt.month
# df['cmonth'] = df.apply(func, axis=1)
# df['cohort month'] = df
def get_period_and_year(date):
  date_obj = pd.to_datetime(date)
  month = date_obj.month
  year = date_obj.year
  if month <= 6:
    return f"{year}-H1"  # Year + First half (H1)
  else:
    return f"{year}-H2"

df['quarters'] = df['cohort date'].apply(get_period_and_year)
    
```

```python
def diff(col1,col2):
    curr_yr = df[col1].dt.year
    curr_mon = df[col1].dt.month
    cohort_yr = df[col2].dt.year
    cohort_mon = df[col2].dt.month
    return (curr_yr - cohort_yr)*2 + (curr_mon - cohort_mon)//6

df['cohort_index'] = diff('Order Date', 'cohort date')
```

```python
df.head()
```

```python
df['cohort_index'].value_counts()
```

```python
cohort_counts = df.groupby(['quarters', 'cohort_index'])['EmailID'].nunique()

cohort_counts_df = cohort_counts.to_frame().rename(columns={'EmailID':'users'}) \
                                .sort_values(by=['quarters'])
```

```python
cohort_counts_df.head(10)
```

```python
cohort_counts_df['retention_rate'] = cohort_counts_df['users'] / cohort_counts_df.groupby(["quarters"])["users"].transform("max")
```

```python
cohorts_pivot = cohort_counts_df.pivot_table(index='quarters', columns = 'cohort_index', values = 'retention_rate')
```

```python
cohorts_pivot
```

```python
# import seaborn as sns
# import matplotlib.pyplot as plt

plt.figure(figsize = (12, 10))
sns.heatmap(data = cohorts_pivot,
            annot = True,
            fmt = '.2%',
            vmin = 0.0,
            vmax = 1.0,
            cmap = 'BuGn')
```
